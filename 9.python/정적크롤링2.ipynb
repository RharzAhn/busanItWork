```python
locations=soup.find_all('location')
for loc in locations:
    city=loc.find('city').string
    print(city)
    datas=loc.find_all('data')
    for data in datas:
        mode=data.find('mode').text
        datetime=data.find('tmef').text # 문서상 tmEf 인데 -> 찾을때는 소문자로 tmef 로 해야되더라 한번씩 테스트해보기
        wf=data.find('wf').text
        print(mode, datetime, wf)
```

```python
locations2=soup.select('location')
for loc in locations2:
    city = loc.select('city')[0].text #select 결과는 무조건 list 그래서 뽑아올려면 [0] 배열을 지정해주어야한다.
    print(city)
    datas=loc.select('data')
    for data in datas:
        mode=data.select('mode')[0].text
        datetime=data.find('tmef').text # 문서상 tmEf 인데 -> 찾을때는 소문자로 tmef 로 해야되더라 한번씩 테스트해보기
        wf=data.find('wf').text
        print(mode, datetime, wf)
```

```python
base="https://ko.wikisource.org"
url1=base+"/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res=req1.urlopen(url1)
res1=req2.get(url1).text
soup=BeautifulSoup(res,'html.parser')
li_list = soup.select('div.mw-parser-output>ul>li')
for li in li_list:
    a= li.select_one('a')
    print('*',a.string)
    sub_li_list = li.find_all('li')
    if sub_li_list != None:
        url2=base+a['href']
        content_out(url2)
    for sub_li in sub_li_list:
        sub_a=sub_li.a
        url2=base+sub_a['href']
        #print(url2)
        print("-",sub_a.string)
        content_out(url2)
```

```python
def content_out(url):
    res=req1.urlopen(url)
    bs=BeautifulSoup(res, 'html.parser')
    content=bs.select('div.mw-parser-output > div>p') #one 은 하나만 가져온다)
    if len(content)>0:
        print(content[0].text)
```

```python
def content_print(url):
    res=req2.get(url).text
    bs=BeautifulSoup(res, 'html.parser')
    content=bs.select('div.poem>p')
    for c in content:
        print(c.text)
```
