```python
import urllib.request as req1
import requests
from bs4 import BeautifulSoup
import pandas as pd
import pymysql
```

```python
url="http://www.yes24.com/Product/Search?domain=ALL&query=python"
res=req1.urlopen(url)
bs=BeautifulSoup(res,'html.parser')
books=bs.select('#yesSchList > li')
# for book in book_list:
#     title=book.select('div > div.item_info > div.info_row.info_name > a.gd_name')
#     test=book.select('div > div.item_info > div.info_row.info_pubGrp')  #뒤에 저자,퍼블리셔등이 나뉘는데 한번에 가져와도 되고, 한단계 더 나눠서 하나씩 가져와도 됨
#     #print(book)
#     print(title[0].text)
#     print(test[0].text)
```

```python
def image_save(url,i): #i는 숫자
    save_name='data/book_images/image'+str(i)+'.jpg'
    req1.urlretrieve(url,save_name)
    print(save_name,'가 저장됨')
    
```

```python
book_list=[]
i=0
for book in books:
    author=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_auth > a') #저자
    pub=book.select('div > div.item_info > div.info_row.info_pubGrp > span.authPub.info_pub')  #출판사
    #print(book)
    #print(author[0].text)
    #print(pub[0].text)
    b_list=[author[0].text,pub[0].text]
    book_list.append(b_list)
    i+=1
    img=book.select('div > div.item_img > div.img_canvas > span > span > a > em > img')
    img_src=img[0]['data-original']
    image_save(img_src,i)
#print(book_list)
df=pd.DataFrame(book_list, columns=['author','pub'])
df
```
