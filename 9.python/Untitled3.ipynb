```python
#import urllib.request
import urllib.request as req
from urllib.request import urlopen
```

```python
#res = urllib.request.urlopen("http://www.naver.com/") #임포트 첫줄을 사용할때 이용
#res = req.urlopen("http://www.naver.com/") #임포트 둘째줄을 사용할때 이용
res = urlopen("http://www.naver.com/") #임포트 셋째줄 사용할때 이용
print(type(res))
print(res.status)
print(res.version)
print(res.msg)

res_headers=res.getheaders()
print("[header 정보]===========================")
for header in res_headers:
    print(header)
    
print("[body 정보]===========================")
print(res.read().decode('utf=8'))
```
