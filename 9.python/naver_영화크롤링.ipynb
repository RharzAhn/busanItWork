```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url='https://movie.naver.com/movie/running/current.naver'
base='http://movie.naver.com/'
res=requests.get(url).content
soup=bs(res, 'html.parser')
#print(soup)
movies=soup.select('div.lst_wrap > ul > li')
#print(movies[0].text)
movie_list=[]

for movie in movies:
    title=movie.select('dl > dt > a')[0].text #셀렉트는 값이 한개라도 리스트로 가져옴 그래서 확인할라면 [0].text붙여줘야함
    rated=movie.select_one('dl > dt > span').text #셀렉트원은 [0]없이 텍스트만 붙여줘도됨
    point=movie.select_one('dl > dd.star > dl > dd > div > a > span.num').text
    print(title,rated,point)
```
